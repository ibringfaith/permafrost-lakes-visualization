# Permafrost and Permafrost Lakes: What are they?

## Permafrost

Permafrost is a permanently frozen layer on or under Earth's surface. It consists of soil, gravel, and sand, usually bound together by ice. Permafrost usually remains at or below 0°C (32ºF) for at least two years.

Permafrost can be found on land and below the ocean floor. It is found in areas where temperatures rarely rise above freezing whether it is at high latitudes or high elevation. This means permafrost is often found in Arctic regions such as Greenland, the U.S. state of Alaska, Canada, Russia, China, and Eastern Europe.

Find out more:
https://education.nationalgeographic.org/resource/permafrost/

## Thermokarst Lakes

### Formation
Degradation of excess ground ice causes ground subsidence
and accumulation of surface water (pond formation). With
continued permafrost thaw, the pond laterally expands into a
lake as the shoreline erodes and collapses. The lake may also
expand vertically as a result of downward talik expansion (unfrozen
soil surrounded by permafrost) and ground subsidence following
thaw of ice-rich permafrost. With sustained permafrost thaw,
taliks expand vertically and laterally, and lake water may flow above
the permafrost or within the permafrost in intrapermafrost taliks.
If the talik extends through the entire permafrost thickness (most
likely where the permafrost is thin, such as the southern extent
of the discontinuous permafrost zone), the lake may expand (if
hydraulic pressure gradients point upward) or drain (if hydraulic
pressure gradients point downward). Increased nutrient flow
from groundwater upwelling may support new aquatic vegetation
such as floating mats. In both permafrost regions (continuous
and discontinuous), lake infilling may occur as thaw-induced
slope failure introduces sediments to the lake, or as a result of
enhanced sediment transport from increasing terrestrial–aquatic
connectivity. Surface lake drainage may occur when a new outlet
forms as a result of melting excess ground ice or an extreme
precipitation event that overtops the lake edge and initiates
mechanical and thermal erosion.  (from nature geoscience Elizabeth E. Webb & Anna K. Liljedahl: https://www.nature.com/articles/s41561-023-01128-z)  
These lakes can exceed 40% of the land surface in some Arctic lowlands and provide feeding and nesting sites for wildlife as well as water sources, hunting and fishing for local communities.

### Global warming impact
On millennial timescales, natural fluctuations in the number and size of lakes are common but decadal scale studies show that climate change is shifting these dynamic systems across a hydrological threshold, towards landscape-scale wetting (lake formation and/or expansion) or drying (lake drainage, desiccation or infilling). Shrinking lake area extent, for example, could mean a reduction in liquid water sources for villagers and habitat for fish and other aquatic organisms whereas Thermokarst lake initiation and expansion increase methane and carbon dioxide fluxes to the atmosphere.

# Cleaning the data

For the coming work, only a few columns are useful. We'll drop the ones we won't use to optimize memory usage using https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html. Only keep the following columns: Shore_len,	Lake_area,	Longitude,	Latitude,	Country,	Mountain,	Endorheic,	Reservoir, GOODD,	SWORD,	Glacier,	Permafrost.  
A little work must be done with the different lake type columns:
- The permafrost and Endorheic columns should be 0 or 1. All non-zero values should be casted to 1.
- The GOODD and Reservoir categories are similar. The value of the Reservoir column should be set to 1 when the GOODD value is 1. When that's done, the GOODD column can be dropped.
- Some lakes don't belong in any categories. For future visulization purposes, a new category 'Other' should be added and the value should be set to 1 for all lakes that don't belong to any other category.

```python
selected_cols = ['Shore_len', 'Lake_area', 'Longitude', 'Latitude', 'Country', 'Mountain', 'Endorheic', 'Reservoir', 'GOODD', 'SWORD', 'Glacier', 'Permafrost']
GLRSED = GLRSED[selected_cols]
GLRSED.head()
```

```python
# permafrost and Endorheic columns should be 0 or 1. All non-zero values should
# be casted to 1
GLRSED['Permafrost'] = (GLRSED['Permafrost'] > 0).astype(int)
GLRSED['Endorheic'] = (GLRSED['Endorheic'] > 0).astype(int)
```

```python
# GOODD and Reservoir columns are similar
GLRSED['Reservoir'] = GLRSED['Reservoir'].where(GLRSED['GOODD'] != 1, 1)
GLRSED = GLRSED.drop(columns=['GOODD'])
```

```python
# add 'Other' category
GLRSED['Other'] = 0
GLRSED.loc[
    (GLRSED['Permafrost'] == 0) &
    (GLRSED['Endorheic'] == 0) &
    (GLRSED['Glacier'] == 0) &
    (GLRSED['SWORD'] == 0) &
    (GLRSED['Reservoir'] == 0) &
    (GLRSED['Mountain'] == 0),
    'Other'
] = 1
```

```python
GLRSED.head()
```

Let's define the different categories of lakes present in the dataset:
- **Mountain**: the lake is located in a mountainous region and is located at high altitude
- **Endhoreic**: there is no water flowing out of the lake
- **Reservoir**: a reservoir is a manmade lake that is created when a dam is built on a river
- **SWORD**: the lake is located on a river

## Lake Types in North America
The first step of this study will be to identify the importance of permafrost lakes in North America. Let's plot a bar chart counting the number and summing the surface of lakes in each category.  
For some of the following visualizations, you might have to transform the dataframe first. Here are the pandas documentation pages for the key transformation you might need to complete the project:
- pd.DataFrame.melt(): https://pandas.pydata.org/docs/reference/api/pandas.melt.html

```python
# create a new dataframe summarizing the data for each lake type
long_data = GLRSED.melt(id_vars=['Country', 'Lake_area', 'Shore_len', 'Longitude', 'Latitude'], var_name='LakeType', value_name='IsType')
long_data = long_data[long_data['IsType'] == 1]
long_data.head()
```

```python
# Plotting bar charts side by side: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharey=None)

# Number of lakes of each type
axes[0].bar(x=North_America_summary['LakeType'], height=North_America_summary['SumLake'], color='cornflowerblue')
axes[0].set_title("Count by Lake Type")
axes[0].set_xlabel("Lake Type")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis='x', rotation=45)

# Area of lakes
axes[1].bar(x=North_America_summary['LakeType'], height=North_America_summary['SumArea'], color='cornflowerblue')
axes[1].set_title("Area by Lake Type")
axes[1].set_xlabel("Lake Type")
axes[1].set_ylabel("Area")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Dependence by country

Our dataset contains the lake data for countries in Central and North America. Let's try to identify a mix of lakes for each of these countries.
- Using pandas, let's create a dataframe summarizing the data by countries
- We'll add a column witht the percentage of the number of lakes of said countries from each category. Some lakes are part of more than one category and will counted more than once doing this. The bars of some countries will therefore exceed 100%.
- Let's visualize this by creating a horizontal stacked bar chart using the altair library (much easier than using matplotlib): https://altair-viz.github.io/gallery/horizontal_stacked_bar_chart.html

```python
# Calculate total count and total area by country
total = GLRSED.groupby('Country').agg(TotalCount=('Lake_area', 'count'), TotalArea=('Lake_area', 'sum')).reset_index()
total.head()
```

```python
long_data
```

```python
# dataframe summarizing the data by country.
long_data = GLRSED.melt(
    id_vars=['Country', 'Lake_area'],
    var_name='LakeType',
    value_name='IsType'
)
long_data = long_data[long_data['IsType'] == 1]
country_summary = long_data.groupby(['Country', 'LakeType']).agg(SumLake=('IsType', 'sum'), SumArea=('Lake_area', 'sum')).reset_index()
country_summary.head()
```

```python
country_summary = pd.merge(country_summary, total, on='Country', how='left')
country_summary.head()
```

```python
# calculate percentage of lakes and area for each lake type in each country
country_summary['percent_nLake'] = country_summary['SumLake'] / country_summary['TotalCount'] * 100
country_summary['percent_area'] = country_summary['SumArea'] / country_summary['TotalArea'] * 100
country_summary.head()
```

```python
# plot a horizontal stacked bar chart using altair
import altair as alt

alt.Chart(country_summary).mark_bar().encode(
    x=alt.X('percent_nLake:Q').title('Percentage of Lakes'),
    y=alt.Y('Country:N', sort='-x'),
    color=alt.Color('LakeType:N', legend=alt.Legend(title="Lake Type")),
    tooltip=['Country', 'LakeType', 'percent_nLake']
).properties(
    title='Percentage of Lakes by Country',
    width=600,
    height=400
)
```

The permafrost lakes are in Greenland (92% of lakes), the United States of America (26% of lakes), and Canada (81% of lakes).

## Dependence by Latitude

Intuitively, we said that permafrost lakes were located at higher latitudes. Let's verify this.
- Bin latitude into 2-degrees interval. Latitudes in our dataset vary from 0° to 90°.
- Plot the number of lakes of each category in each latitude bin in a vertical stacked bar chart  

If you need a little reminder on Latitude and Longitude, check out: https://www.britannica.com/science/latitude

```python
import numpy as np

# Binning latitude into 2-degree intervals starting from 0°
GLRSED['LatBin'] = pd.cut(GLRSED['Latitude'], bins=np.arange(0, 91, 2))

GLRSED.head()
```

```python
# aggregate data by Latitude Bins
latitude_summary = GLRSED.groupby(['LatBin']).agg(CountPermafrost= ('Permafrost', 'sum'), CountGlacier= ('Glacier', 'sum'), CountSWORD= ('SWORD', 'sum'), CountMountain= ('Mountain', 'sum'),
                                             CountReservoir= ('Reservoir', 'sum'), CountEndorheic= ('Endorheic', 'sum'), CountOther= ('Other', 'sum')).reset_index()
latitude_summary.head()
```

```python
plot = latitude_summary.plot(x='LatBin', kind='bar', stacked=True, figsize=(10, 6))

plt.title('Lake Type by Latitude')
plt.xlabel('Latitude Bin')
plt.ylabel('Count Per Lake Type')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.setp(plot.axes.get_xticklabels()[::2], visible=False)
plt.legend(title='Lake Type')

plt.tight_layout()
plt.show()
```

## Heatmap of permafrost lakes

Our last plot will be the most complex one. Building on our plots by countries and latitude, we willl now plot a map of North America superposed with a heatmap of the number of permafrost lakes.
A few hints:
- A world map is available in the geopandas library using .datasets.get_path('naturalearth_lowres'). You should filter the desired countries.
- Bin our clean dataset based on latitude and longitude with 1-degree intervals and group the dataset based on these bins.
- To plot the heatmap we are going to use the imshow funtion from matplotlib. In order to do so you should create a matrix with rows being latitude bins, columns being longitude bins and filled with the number of permafrost lakes in square ixj.


```python
uploaded_two = files.upload()
```

```python
import geopandas as gpd

world = gpd.read_file('ne_110m_admin_0_countries.shp')
world.head()
```

```python
north_america_countries = ['United States of America', 'Canada', 'Greenland', 'Mexico', 'Belize', 'Costa Rica',
                            'Cuba', 'Dominican Republic', 'El Salvador', 'Guatemala', 'Haiti',
                            'Honduras', 'Jamaica', 'Martinique', 'Nicaragua',
                            'Panama', 'Puerto Rico', 'Trinidad and Tobago']

north_america = world[world['NAME'].isin(north_america_countries)]

north_america.head()
```

```python
# Bin longitude and latitude using 1-degree intervals and summarize the data by longittude and latitude bins
GLRSED['LonBin'] = GLRSED['Longitude'].astype(int)
GLRSED['LatBin'] = GLRSED['Latitude'].astype(int)
GLRSED_lat_lon_bins = GLRSED.groupby(['LatBin', 'LonBin']).agg(
    TotalLakes=('Lake_area', 'count'),
    PermafrostLakes=('Permafrost', 'sum')
).reset_index()
GLRSED_lat_lon_bins.head()
```

```python
# Add a column with the percentage of lakes that are permafrost lakes.
GLRSED_lat_lon_bins['percent_nLakes'] = (GLRSED_lat_lon_bins['PermafrostLakes'] / GLRSED_lat_lon_bins['TotalLakes']) * 100
GLRSED_lat_lon_bins.head()
```

```python
# make a matrix from dataframe
permafrost_matrix = GLRSED_lat_lon_bins.pivot(index='LatBin', columns='LonBin', values='percent_nLakes').replace(0, np.nan)[::-1]

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))

# use imshow to plot the heatmap of the permafrost lakes
im=ax.imshow(permafrost_matrix, extent=[
    permafrost_matrix.columns.min(), permafrost_matrix.columns.max(),
    permafrost_matrix.index.min(), permafrost_matrix.index.max()
], cmap='coolwarm', interpolation='nearest')

# Plot North American countries boundaries
north_america.boundary.plot(ax=ax, linewidth=2, color='black')

# Set title and labels. Add colorbar with https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.colorbar.html
plt.title('Percentage of Permafrost Lakes by Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(im, ax=ax, label='% Permafrost Lakes')
plt.tight_layout()
plt.show()
```
